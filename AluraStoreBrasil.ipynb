### Importa√ß√£o dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. An√°lise do faturamento



```python
# üí∞ Faturamento Total de Cada Loja

fat1 = loja['Pre√ßo'].sum()
fat2 = loja2['Pre√ßo'].sum()
fat3 = loja3['Pre√ßo'].sum()
fat4 = loja4['Pre√ßo'].sum()

print("========= Faturamento Total =========")
print(f"Loja 1: R${fat1:,.2f}")
print(f"Loja 2: R${fat2:,.2f}")
print(f"Loja 3: R${fat3:,.2f}")
print(f"Loja 4: R${fat4:,.2f}")
print("====================================")
```

# 2. Vendas por Categoria

```python
cat1 = loja["Categoria do Produto"].value_counts()
cat2 = loja2["Categoria do Produto"].value_counts()
cat3 = loja3["Categoria do Produto"].value_counts()
cat4 = loja4["Categoria do Produto"].value_counts()

print("Loja 1 - Categorias mais vendidas:")
print(cat1, "\n")

print("Loja 2 - Categorias mais vendidas:")
print(cat2, "\n")

print("Loja 3 - Categorias mais vendidas:")
print(cat3, "\n")

print("Loja 4 - Categorias mais vendidas:")
print(cat4, "\n")
```

# 3. M√©dia de Avalia√ß√£o das Lojas


```python
media1 = loja['Avalia√ß√£o da compra'].mean()
media2 = loja2['Avalia√ß√£o da compra'].mean()
media3 = loja3['Avalia√ß√£o da compra'].mean()
media4 = loja4['Avalia√ß√£o da compra'].mean()

print("== M√©dia de Avalia√ß√µes ==")
print(f"Loja 1: {media1:.2f}")
print(f"Loja 2: {media2:.2f}")
print(f"Loja 3: {media3:.2f}")
print(f"Loja 4: {media4:.2f}")
print("=========================")
```

# 4. Produtos Mais e Menos Vendidos

```python
# Loja 1
mais_vendido1 = loja['Produto'].value_counts().idxmax()
menos_vendido1 = loja['Produto'].value_counts().idxmin()

# Loja 2
mais_vendido2 = loja2['Produto'].value_counts().idxmax()
menos_vendido2 = loja2['Produto'].value_counts().idxmin()

# Loja 3
mais_vendido3 = loja3['Produto'].value_counts().idxmax()
menos_vendido3 = loja3['Produto'].value_counts().idxmin()

# Loja 4
mais_vendido4 = loja4['Produto'].value_counts().idxmax()
menos_vendido4 = loja4['Produto'].value_counts().idxmin()

print("====================== Produtos Mais e Menos Vendidos ======================")
print(f"Loja 1 ‚Üí Mais vendido: {mais_vendido1} | Menos vendido: {menos_vendido1}")
print(f"Loja 2 ‚Üí Mais vendido: {mais_vendido2} | Menos vendido: {menos_vendido2}")
print(f"Loja 3 ‚Üí Mais vendido: {mais_vendido3} | Menos vendido: {menos_vendido3}")
print(f"Loja 4 ‚Üí Mais vendido: {mais_vendido4} | Menos vendido: {menos_vendido4}")
print("============================================================================")
```

# 5. Frete M√©dio por Loja

```python
frete1 = loja['Frete'].mean()
frete2 = loja2['Frete'].mean()
frete3 = loja3['Frete'].mean()
frete4 = loja4['Frete'].mean()

print("==== Frete M√©dio por Loja ====")
print(f"Loja 1: R${frete1:.2f}")
print(f"Loja 2: R${frete2:.2f}")
print(f"Loja 3: R${frete3:.2f}")
print(f"Loja 4: R${frete4:.2f}")
print("==============================")
```

# Gr√°fico de Colunas ‚Äî Faturamento Total por Loja
Mostra o quanto cada loja faturou (√≥timo para comparar visualmente).

```python
import matplotlib.pyplot as plt

faturamentos = [fat1, fat2, fat3, fat4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8,5))
plt.bar(lojas, faturamentos, color='#2ecc71')
plt.title('Faturamento Total por Loja', fontsize=14, fontweight='bold')
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.gca().get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda x, _: f'{int(x):,}'.replace(',', '.'))
)

plt.tight_layout()
plt.show()
```

# Gr√°fico de Pizza ‚Äî Distribui√ß√£o das Categorias
Mostra quais categorias mais vendem em uma loja espec√≠fica.

```python
categorias_loja4 = loja['Categoria do Produto'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(categorias_loja4, labels=categorias_loja4.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribui√ß√£o das Categorias de Produtos - Loja 4')
plt.show()
```

# Gr√°fico de √Årea ‚Äî Frete M√©dio por Loja
Mostra a diferen√ßa de custo m√©dio de frete entre as lojas.

```python
fretes = [frete1, frete2, frete3, frete4]

plt.figure(figsize=(8,5))
plt.fill_between(lojas, fretes, color='lightgreen', alpha=0.6)
plt.plot(lojas, fretes, color='green', marker='o')
plt.title('Frete M√©dio por Loja')
plt.xlabel('Lojas')
plt.ylabel('Frete M√©dio (R$)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
